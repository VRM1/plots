```python
%matplotlib notebook   
# %matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import rc
import numpy as np
import pandas as pd
```

```python
# load variables and data
figure_nam = 'my_fig.png'
n_col = 2 #No. of column in legend
loc = 2 #Location for legend
lc='dataset/'
# load data
fil=[lc+'priors_restr.csv',lc+'priors_hobbit.csv',lc+'priors_civilwar.csv',lc+'priors_camera.csv']
df1,df2,df3,df4=pd.read_csv(fil[0]),pd.read_csv(fil[1]),pd.read_csv(fil[2]),pd.read_csv(fil[3])
data_x=df1['Unnamed: 0'] # 'names for each group of plots.'
# load data
data1= [df1['T-50'],df1['T-100'],df1['T-150']]
data2= [df2['T-50'],df2['T-100'],df2['T-150']]
data3= [df3['T-50'],df3['T-100'],df3['T-150']]
data4= [df4['T-50'],df4['T-100'],df4['T-150']]
```

```python
def bar_plot2(data_x,data_y1,data_y2,data_y3,data_y4):
    font = {'family' : 'Serif',
        'weight' : 'normal',
        'size'   : 14}
    rc('font', **font)
    
    #set color of the bar in the plot
    c2='#D1D0CE'
    c3='#777777'
    c4='#000000'
    
    #To plot
    fig = plt.figure(num=1,figsize=(7,8))
    ax = fig.add_subplot(121)
    list_vals1=np.array(data_y1[0])
    list_vals2=np.array(data_y1[1])
    list_vals3=np.array(data_y1[2])
            
    N = len(list_vals1)      
    ind = np.arange(N)  # the x locations for the groups
    width = 0.2       # the width of the bars
    
    b1 = ax.bar(ind, list_vals1, width, color=c2)
    b2 = ax.bar(ind+0.2, list_vals2, width, color=c3)
    b3 = ax.bar(ind+0.4, list_vals3, width, color=c4)
    
    # add some text for labels, title and axes ticks
    ax.set_ylabel('p@10',fontsize=18)
    ax.set_xlabel('Restr',fontsize=18)
    ax.set_xticks(ind+width*2)
    ax.set_xticklabels(data_x)
    ax.legend((b1,b2,b3), ('T-50','T-100','T-150'),ncol=n_col,prop={'size':10},frameon=True,loc=loc)

    #2nd plot
    ax = fig.add_subplot(122)
    list_vals1=np.array(data_y2[0])
    list_vals2=np.array(data_y2[1])
    list_vals3=np.array(data_y2[2])
            
    N = len(list_vals1)      
    ind = np.arange(N)  # the x locations for the groups
    width = 0.2       # the width of the bars
    
    b1 = ax.bar(ind, list_vals1, width, color=c2)
    b2 = ax.bar(ind+0.2, list_vals2, width, color=c3)
    b3 = ax.bar(ind+0.4, list_vals3, width, color=c4)
    
    # add some text for labels, title and axes ticks
    ax.set_ylabel('P@10',fontsize=18)
    ax.set_xlabel('Hobbit',fontsize=18)
    ax.set_xticks(ind+width*2)
    ax.set_xticklabels(data_x)
    ax.legend((b1,b2,b3), ('T-50','T-100','T-150'),ncol=n_col,prop={'size':10},frameon=True,loc=loc)
    
    # add some text for labels, title and axes ticks
    ax.set_ylabel('P@10',fontsize=18)
    ax.set_xlabel('Camera',fontsize=18)
    ax.set_xticks(ind+width*2)
    ax.set_xticklabels(data_x)
    ax.legend((b1,b2,b3), ('T-50','T-100','T-150'),ncol=n_col,prop={'size':10},frameon=True,loc=loc)
    plt.subplots_adjust(bottom = 0.1,top = 0.95,wspace = 0.2,hspace = 0.3 )
    plt.savefig(figure_nam,orientation='portrait',bbox_inches='tight',dpi=300)
#     plt.show()
bar_plot2(data_x,data1,data2,data3,data4)
```
