```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import rc
import numpy as np
import pandas as pd
```

```python
# load variables and data
figure_nam = 'my_fig.png'
lc='dataset/'
#No of topics
tops=[5,10,30,50,100,150,200]
# load data
fil=[lc+'coherence_restr.csv',lc+'coherence_hobbit.csv',lc+'coherence_civilwar.csv',lc+'coherence_camera.csv']
df1,df2,df3,df4=pd.read_csv(fil[0]),pd.read_csv(fil[1]),pd.read_csv(fil[2]),pd.read_csv(fil[3])
```

```python
# load variables and data
figure_nam = 'my_fig.png'
#No of topics
tops=df1['Topics']
# load data
data1= [df1['LDA'],df1['MGLDA'],df1['M-ASUM'],df1['TTM'],df1['APSUM']]
data2= [df2['LDA'],df2['MGLDA'],df2['M-ASUM'],df2['TTM'],df2['APSUM']]
data3= [df3['LDA'],df3['MGLDA'],df3['M-ASUM'],df3['TTM'],df3['APSUM']]
data4= [df4['LDA'],df4['MGLDA'],df4['M-ASUM'],df4['TTM'],df4['APSUM']]
```

```python
def dcg_new(data_x,data1,data2,data3,data4,figure_nam,x_axis,y_lim=1,n_col=2,loc=2):
    font = {'family' : 'Serif',
        'weight' : 'normal',
        'size'   : 14}
    rc('font', **font)
    fig = plt.figure(num=1,figsize=(12,8))
    lb=['LDA', 'MGLDA', 'M-ASUM','TTM','APSUM']
    x_label='Topics\n'
    x_tick_label=[10,30,50,100,150,200]
    
    list_vals1=np.array(data1[0])
    list_vals2=np.array(data1[1])
    list_vals3=np.array(data1[2])
    list_vals4=np.array(data1[3])
    list_vals5=np.array(data1[4])
    
    ax = fig.add_subplot(221)
    rects1 = ax.plot(data_x,list_vals1,linewidth=1, marker='o', color='#7FFF00',label=lb[0])
    rects2 = ax.plot(data_x,list_vals2,linewidth=1, marker='x', color='#FF7F50',label=lb[1])
    rects3 = ax.plot(data_x,list_vals3,linewidth=1, marker='*', color='#DC143C',label=lb[2])
    rects4 = ax.plot(data_x,list_vals4,linewidth=1, marker='s', color='#00008B',label=lb[3])
    rects5 = ax.plot(data_x,list_vals5,linewidth=1, marker='H', color='#8A2BE2',label=lb[4])
    ax.legend(loc=4,ncol=2,prop={'size':10},frameon=True )
    ax.set_xticks(x_tick_label)
#    ax.grid(linestyle='-')
    ax.set_ylim(0,2.0)
    ax.set_ylabel(x_axis,fontsize=18)
    ax.set_xlabel(x_label+'(a) Restr',fontsize=18)
    
    
    list_vals1=np.array(data2[0])
    list_vals2=np.array(data2[1])
    list_vals3=np.array(data2[2])
    list_vals4=np.array(data2[3])
    list_vals5=np.array(data2[4])
    
    ax = fig.add_subplot(222)
    rects1 = ax.plot(data_x,list_vals1,linewidth=1, marker='o', color='#7FFF00',label=lb[0])
    rects2 = ax.plot(data_x,list_vals2,linewidth=1, marker='x', color='#FF7F50',label=lb[1])
    rects3 = ax.plot(data_x,list_vals3,linewidth=1, marker='*', color='#DC143C',label=lb[2])
    rects4 = ax.plot(data_x,list_vals4,linewidth=1, marker='s', color='#00008B',label=lb[3])
    rects5 = ax.plot(data_x,list_vals5,linewidth=1, marker='H', color='#8A2BE2',label=lb[4])
    ax.legend(loc=4,ncol=2,prop={'size':10},frameon=True )
    ax.set_xticks(x_tick_label)
    ax.set_ylim(0,2.5)
    ax.set_ylabel(x_axis,fontsize=18)
    ax.set_xlabel(x_label+'(b) Hobbit',fontsize=18)
    
    list_vals1=np.array(data3[0])
    list_vals2=np.array(data3[1])
    list_vals3=np.array(data3[2])
    list_vals4=np.array(data3[3])
    list_vals5=np.array(data3[4])
    
    ax = fig.add_subplot(223)
    rects1 = ax.plot(data_x,list_vals1,linewidth=1, marker='o', color='#7FFF00',label=lb[0])
    rects2 = ax.plot(data_x,list_vals2,linewidth=1, marker='x', color='#FF7F50',label=lb[1])
    rects3 = ax.plot(data_x,list_vals3,linewidth=1, marker='*', color='#DC143C',label=lb[2])
    rects4 = ax.plot(data_x,list_vals4,linewidth=1, marker='s', color='#00008B',label=lb[3])
    rects5 = ax.plot(data_x,list_vals5,linewidth=1, marker='H', color='#8A2BE2',label=lb[4])
    ax.legend(loc=4,ncol=2,prop={'size':10},frameon=True )
    ax.set_xticks(x_tick_label)
    ax.set_ylim(0,2.5)
    ax.set_ylabel(x_axis,fontsize=18)
    ax.set_xlabel(x_label+'(c) Civil War',fontsize=18)
    
    list_vals1=np.array(data4[0])
    list_vals2=np.array(data4[1])
    list_vals3=np.array(data4[2])
    list_vals4=np.array(data4[3])
    list_vals5=np.array(data4[4])
    
    ax = fig.add_subplot(224)
    rects1 = ax.plot(data_x,list_vals1,linewidth=1, marker='o', color='#7FFF00',label=lb[0])
    rects2 = ax.plot(data_x,list_vals2,linewidth=1, marker='x', color='#FF7F50',label=lb[1])
    rects3 = ax.plot(data_x,list_vals3,linewidth=1, marker='*', color='#DC143C',label=lb[2])
    rects4 = ax.plot(data_x,list_vals4,linewidth=1, marker='s', color='#00008B',label=lb[3])
    rects5 = ax.plot(data_x,list_vals5,linewidth=1, marker='H', color='#8A2BE2',label=lb[4])
    ax.legend(loc=4,ncol=2,prop={'size':10},frameon=True )
    ax.set_xticks(x_tick_label)
    ax.set_ylim(0,2.0)
    ax.set_ylabel(x_axis,fontsize=18)
    ax.set_xlabel(x_label+'(d) Camera',fontsize=18)
    
    plt.subplots_adjust(bottom = 0.1,top = 0.9,wspace = 0.2,hspace = 0.4 )
#     plt.show()
    plt.savefig(figure_nam,orientation='portrait',bbox_inches='tight',dpi=300)
dcg_new(tops, data1, data2, data3, data4, figure_nam, x_axis='Coherence', y_lim=2, n_col=2)
```
