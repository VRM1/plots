```python
def dcg_new(data_x,data1,data2,data3,data4,figure_nam,x_axis,y_lim=1,n_col=2,loc=2):
    font = {'family' : 'Serif',
        'weight' : 'normal',
        'size'   : 14}
    rc('font', **font)
    fig = plt.figure(num=1,figsize=(12,8))
    lb=['LVAE', 'CF-LVAE','PMF','Random']
    x_label='K\n'
#     x_tick_label=[100,200,400,600,800]
    x_tick_label=data_x
    list_vals1=np.array(data1[0])
    list_vals2=np.array(data1[1])
    list_vals3=np.array(data1[2])
    list_vals4=np.array(data1[3])
    
    ax = fig.add_subplot(221)
    rects1 = ax.plot(data_x,list_vals1,linewidth=1, marker='o', color='#7FFF00',label=lb[0])
    rects2 = ax.plot(data_x,list_vals2,linewidth=1, marker='x', color='#FF7F50',label=lb[1])
    rects3 = ax.plot(data_x,list_vals3,linewidth=1, marker='*', color='#DC143C',label=lb[2])
    rects4 = ax.plot(data_x,list_vals4,linewidth=1, marker='s', color='#00008B',label=lb[3])

    ax.legend(loc=1,ncol=2,prop={'size':10},frameon=True )
    ax.set_xticks(x_tick_label)
#    ax.grid(linestyle='-')
    ax.set_xlim(10,800)
    ax.set_ylim(0.000001,1)
    plt.yscale('log')
    ax.set_ylabel(x_axis,fontsize=18)
    ax.set_xlabel(x_label+"(a) Men's Clothing (Substitute)" ,fontsize=18)
    
    
    list_vals1=np.array(data2[0])
    list_vals2=np.array(data2[1])
    list_vals3=np.array(data2[2])
    list_vals4=np.array(data2[3])

    
    ax = fig.add_subplot(222)
    rects1 = ax.plot(data_x,list_vals1,linewidth=1, marker='o', color='#7FFF00',label=lb[0])
    rects2 = ax.plot(data_x,list_vals2,linewidth=1, marker='x', color='#FF7F50',label=lb[1])
    rects3 = ax.plot(data_x,list_vals3,linewidth=1, marker='*', color='#DC143C',label=lb[2])
    rects4 = ax.plot(data_x,list_vals4,linewidth=1, marker='s', color='#00008B',label=lb[3])

    ax.legend(loc=1,ncol=2,prop={'size':10},frameon=True )
    ax.set_xticks(x_tick_label)
    ax.set_xlim(10,800)
    ax.set_ylim(0.000001,1)
    plt.yscale('log')
    ax.set_ylabel(x_axis,fontsize=18)
    ax.set_xlabel(x_label+"(b) Men's Clothing (Supplement)",fontsize=18)
    
    list_vals1=np.array(data3[0])
    list_vals2=np.array(data3[1])
    list_vals3=np.array(data3[2])
    list_vals4=np.array(data3[3])

    
    ax = fig.add_subplot(223)
    rects1 = ax.plot(data_x,list_vals1,linewidth=1, marker='o', color='#7FFF00',label=lb[0])
    rects2 = ax.plot(data_x,list_vals2,linewidth=1, marker='x', color='#FF7F50',label=lb[1])
    rects3 = ax.plot(data_x,list_vals3,linewidth=1, marker='*', color='#DC143C',label=lb[2])
    rects4 = ax.plot(data_x,list_vals4,linewidth=1, marker='s', color='#00008B',label=lb[3])

    ax.legend(loc=1,ncol=2,prop={'size':10},frameon=True )
    ax.set_xticks(x_tick_label)
    ax.set_xlim(10,800)
    ax.set_ylim(0.000001,1)
    plt.yscale('log')
    ax.set_ylabel(x_axis,fontsize=18)
    ax.set_xlabel(x_label+'(c) Electronics (Substitute)',fontsize=18)
    
    list_vals1=np.array(data4[0])
    list_vals2=np.array(data4[1])
    list_vals3=np.array(data4[2])
    list_vals4=np.array(data4[3])

    
    ax = fig.add_subplot(224)
    rects1 = ax.plot(data_x,list_vals1,linewidth=1, marker='o', color='#7FFF00',label=lb[0])
    rects2 = ax.plot(data_x,list_vals2,linewidth=1, marker='x', color='#FF7F50',label=lb[1])
    rects3 = ax.plot(data_x,list_vals3,linewidth=1, marker='*', color='#DC143C',label=lb[2])
    rects4 = ax.plot(data_x,list_vals4,linewidth=1, marker='s', color='#00008B',label=lb[3])

    ax.legend(loc=1,ncol=2,prop={'size':10},frameon=True )
    ax.set_xticks(x_tick_label)
    ax.set_xlim(10,800)
    ax.set_ylim(0.000001,1)
    plt.yscale('log')
    ax.set_ylabel(x_axis,fontsize=18)
    ax.set_xlabel(x_label+'(d) Electronics (Supplement)',fontsize=18)
    
    plt.subplots_adjust(bottom = 0.1,top = 0.9,wspace = 0.2,hspace = 0.4 )
#     plt.show()
    plt.tight_layout()
    plt.savefig(figure_nam,orientation='portrait',bbox_inches='tight',dpi=300)
dcg_new(tops, data1, data2, data3, data4, figure_nam, x_axis='Precision@K', y_lim=2, n_col=2)
```
