```python
%matplotlib notebook
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import rc
import numpy as np
```

```python
# Single bar graph, it can handle upto 4 bars and 3 locs
def bar_plot1(x_data, x_ticks_label, x_label, y_label, legend_names, figure_nam):

    font = {'family' : 'Serif',
        'weight' : 'normal',
        'size'   : 25}
    rc('font', **font)
    colors = ['#38DB0B','#D1D0CE','#777777','#DBB20B']

    plt.figure(num=1,figsize=(14,7))
    ind = np.arange(len(x_data[0]))  # the x locations for the groups
    width = 0.2       # the width of the bars

    for i, vals in enumerate(x_data):
        shift = width * i
        plt.bar(ind+shift, vals, width, color=colors[i], label=legend_names[i], edgecolor='black')

     # add some text for labels, title and axes ticks
    plt.ylabel(y_label)
    plt.xlabel(x_label)
    plt.xticks(ind+width*2,x_ticks_label)
    plt.legend(ncol=1,loc='best')
    plt.savefig(figure_nam, orientation='portrait', bbox_inches='tight', dpi=300)
```

```python
# example for One bar graph
x_tick_labels=['25%','50%','100%'] # 'names for each group of plots.'
x_label = "Elapsed Rounds (%)"
y_label = "Expected Calibration Error"
legend_names = ['MCD-VR',"EN-VR",'BCN-VR','BCN-VR*']
figure_name = "some_x"
x_data=[[0.15,0.26,0.25],[0.1,0.2,0.35],[0.2,0.3,0.45],[0.1,0.38,0.55]]
bar_plot1(x_data, x_tick_labels, x_label, y_label, legend_names, figure_name)
```
